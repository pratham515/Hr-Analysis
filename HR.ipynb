```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hr=pd.read_csv("all_data/HR_Analytics.csv")
hr.head(2)
```

```python
# Finding duplicates 
hr[hr.duplicated()]
```

```python
# removing duplicates 
hr.drop_duplicates(inplace=True)
```

```python
# creating column named atr to store values of yes as 1 and 0 for no 
hr["ATR"]= 0
hr.loc[hr["Attrition"] == "Yes", "ATR"] = 1
```

```python
# crosschecking null values 
hr[hr.isna()].count()
```

```python
hr.info()
```

```python
# saving cleaned data 
hr.to_csv("C:\\Users\\nauti\\Desktop\\python_ducat\\hr_cleaned\\hr_cleaned.csv")
```

# Univariate analysis

```python
# ploting kde on age  
sns.kdeplot(hr["Age"],fill=True,color="red")
plt.title("Age distribution")
plt.show()
```

```python
# attrition
sns.kdeplot(hr["ATR"],fill=True,color="red")
plt.title("Atrrition Dirstribution")
plt.show()
```

```python
# daily rate 
sns.kdeplot(hr["DailyRate"],fill=True,color="red")
plt.title("daily rate Distribution")
plt.show()
```

```python
# YearsAtCompany
sns.kdeplot(hr["YearsAtCompany"],fill=True,color="red")
plt.title("years at company Distribution")
plt.show()
```

# bivariate

```python
sns.barplot(x=hr["AgeGroup"],y=hr["DailyRate"],color='red',alpha=0.7)
plt.title("Hourly rate by age group")
plt.show()
```

```python
piechart= hr.groupby("Department")["ATR"].sum()
label= piechart.index

plt.bar(label,piechart,color="red",alpha=0.7)
plt.title("Attrition by department ")
plt.show()
```

```python
piechart= hr.groupby("EducationField")["ATR"].sum()
label= piechart.index

plt.pie(piechart,labels=label,autopct="%0.1f%%",explode=[0.1,0,0,0,0,0.1])
plt.title("Attrition by Education feild ")
plt.show()
```

```python
val= hr.groupby("Gender")["ATR"].sum()
gen= val.index
sns.barplot(x=gen,y=val,color="red",alpha=0.7)
plt.title("Attrition by gender ")
plt.show()
```

```python
gr= hr.groupby("SalarySlab")["ATR"].sum()
slab= gr.index
sns.barplot(x=slab,y=gr,color="red",alpha=0.7)
plt.title("Attrition by salary slab ")
plt.show()
```

```python
# job profile 
val = hr.groupby("JobRole")["ATR"].sum()
role= val.index


sns.barplot(x=val,y=role,color="red",alpha=0.7)
plt.title("Attrition by job Role ")
plt.show()

```

```python
val=hr.groupby("MaritalStatus")["ATR"].sum()
status= val.index
sns.barplot(y=val,x=status,color="red",alpha=0.7)
plt.title("Attrition by Status ")
plt.show()
```

```python
# job satisfaction in percentage 
pd.crosstab(index=hr["ATR"],columns=hr["JobSatisfaction"],normalize="index")*100
```

```python
dis=hr.groupby("DistanceFromHome")["ATR"].sum()
idx= dis.index

plt.bar(idx,dis,color="red",alpha=0.7)
plt.xlabel("distance")
plt.ylabel("Attrition")
plt.title("Attrition by distance from Office")
plt.show()
```
